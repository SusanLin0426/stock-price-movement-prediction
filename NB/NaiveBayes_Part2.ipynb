```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime

import warnings
warnings.filterwarnings("ignore")
```

### Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
yuanta_data = pd.read_csv('yuanta_vector.csv')
print(yuanta_data.shape)
print(yuanta_data['post_time'].describe())

yuanta_data = yuanta_data[yuanta_data['label'] != 0]
```

```python
print(yuanta_data.shape)
print(yuanta_data['post_time'].describe())
```

```python
yuanta_data.describe()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import ComplementNB
from sklearn.naive_bayes import BernoulliNB 
from sklearn.naive_bayes import CategoricalNB
from sklearn.model_selection import train_test_split


prediction = pd.DataFrame(columns=['gnb_pred', 'mnb_pred', 'cnb_pred', 'bnb_pred', 'catnb_pred', 'true_label'])

X = yuanta_data.drop('label', axis=1)
y = yuanta_data['label']
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.2)

X_train = X_train.drop('post_time', axis=1)
X_validation = X_validation.drop('post_time', axis=1)

gnb = GaussianNB()
gnb.fit(X_train, y_train)
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
cnb = ComplementNB()
cnb.fit(X_train, y_train)
bnb = BernoulliNB()
bnb.fit(X_train, y_train)
catnb = CategoricalNB(min_categories=3)
catnb.fit(X_train, y_train)

for i in range(X_validation.shape[0]):
    print(i)
    gnb_y_pred = gnb.predict(X_validation.iloc[[i],:])
    mnb_y_pred = mnb.predict(X_validation.iloc[[i],:])
    cnb_y_pred = cnb.predict(X_validation.iloc[[i],:])
    bnb_y_pred = bnb.predict(X_validation.iloc[[i],:])
    catnb_y_pred = catnb.predict(X_validation.iloc[[i],:])
    prediction = pd.concat([prediction, pd.DataFrame([[gnb_y_pred, mnb_y_pred, cnb_y_pred, bnb_y_pred, catnb_y_pred, y_validation.iloc[i]]], columns=['gnb_pred', 'mnb_pred', 'cnb_pred', 'bnb_pred', 'catnb_pred', 'true_label'])], ignore_index=True)
```

```python
prediction
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import confusion_matrix
print('GaussianNB')
print("Accuracy: ", accuracy_score(list(prediction['true_label']), list(prediction['gnb_pred'])))
print(confusion_matrix(list(prediction['true_label']), list(prediction['gnb_pred']),labels=[1,-1]))
print("========================")
print('MultinomialNB')
print("Accuracy: ", accuracy_score(list(prediction['true_label']), list(prediction['mnb_pred'])))
print(confusion_matrix(list(prediction['true_label']), list(prediction['mnb_pred'])))
print("========================")
print('ComplementNB')
print("Accuracy: ", accuracy_score(list(prediction['true_label']), list(prediction['cnb_pred'])))
print(confusion_matrix(list(prediction['true_label']), list(prediction['cnb_pred'])))
print("========================")
print('BernoulliNB')
print("Accuracy: ", accuracy_score(list(prediction['true_label']), list(prediction['bnb_pred'])))
print(confusion_matrix(list(prediction['true_label']), list(prediction['bnb_pred'])))
print("========================")
print('CategoricalNB')
print("Accuracy: ", accuracy_score(list(prediction['true_label']), list(prediction['catnb_pred'])))
print(confusion_matrix(list(prediction['true_label']), list(prediction['catnb_pred'])))
```
